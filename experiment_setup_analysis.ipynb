```python
import seaborn as sns
import numpy as np
import pandas as pd
from pprint import pprint
```

```python
data = pd.read_csv("Record_Data\Y025\T03.csv")
print("data columns: ", data.columns)

data = data[["Time", "x", "q1", "q2", "x_dot"]]

sign_change_index = []
current_sign = np.sign(data.iloc[0]["x_dot"])
for index, row in data.iterrows():
    if np.sign(row["x_dot"]) != current_sign: 
        sign_change_index.append(index)
        current_sign = np.sign(row["x_dot"])
```

```python
sns.lineplot(x=data.index, y=data["x_dot"])
```

```python
data["x_dot_abs"] = abs(data.x_dot)
sns.lineplot(x=data.index, y=data["x_dot_abs"])
```

```python
sns.lineplot(x=data.index, y=data["x"])
```

```python
x= [i for i in range(250,1250)]
y = data.iloc[x]["x"]
sns.lineplot(x=x, y=y)
distance = y.max()-y.min()
time_taken = (data.iloc[y.idxmax()]["Time"]-data.iloc[y.idxmin()]["Time"])
avg_vel = distance/time_taken
print("distance: ", distance, "\tm")
print("average velocity: ", avg_vel, "\tm/s")
print("average velocity: ", avg_vel*100, "\tcm/s")
```
